```python
# filtering recipe on page 1
df_filter1 = df.loc[df.index.isin([71])]
df_filter1.head()
```

```python
# detect list type in ners
from ast import literal_eval

df_filter1['ners'] = df_filter1.ners.apply(lambda x: literal_eval(str(x)))
df_filter1.head()
```

```python
# first ner list
ners = df_filter1['ners'].item()
ners
```

```python
# detect list type in words - same doing it for words
from ast import literal_eval

df_filter1['words'] = df_filter1.words.apply(lambda x: literal_eval(str(x)))
df_filter1.head()
```

```python
# first words list
words = df_filter1['words'].item()
words
```

```python
ners, words
```

```python
# index containing out of tag
o_idxs = []
for i, tag in enumerate(ners):
    if tag == 'O':
        o_idxs.append(i)

o_idxs
```
